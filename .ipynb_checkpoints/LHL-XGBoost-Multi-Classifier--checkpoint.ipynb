```python
import numpy as np
import pandas as pd
import xgboost as xgb
```

```python
data_test = pd.read_csv('/Data/test.csv', index_col=0)
data_test
```

```python
data_train = pd.read_csv('train.csv')
data_train
```

```python
#some data cleaning
data_train.duplicated().sum()
```

```python

data_train.isnull().sum()
```

```python
data_test.duplicated().sum()
```

```python
data_test.isnull().sum()
```

```python
#drop duplicated values
data_train.drop_duplicates(keep = 'first', inplace = True)
data_train.shape
```

```python
#data preprocessing
X_train = data_train.drop(['Category', 'Resolution', 'Descript'], axis = 1)
X_train
```

```python
y_train = data_train['Category']
y_train
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
```

```python
y_train
```

```python
#X_train['Descript'] = le.fit_transform(X_train['Descript'])
X_train['PdDistrict'] = le.fit_transform(X_train['PdDistrict'])
```

```python
X_train['DayOfWeek'] = le.fit_transform(X_train['DayOfWeek'])
X_train['Address'] = le.fit_transform(X_train['Address'])
X_train['Dates'] = le.fit_transform(X_train['Dates'])
```

```python
X_train
```

```python
#feature scaling with standard scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)
```

```python
# XGBoost model
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
X_test = data_test
X_test
```

```python
#X_test['Descript'] = le.fit_transform(X_test['Descript'])
X_test['PdDistrict'] = le.fit_transform(X_test['PdDistrict'])
X_test['DayOfWeek'] = le.fit_transform(X_test['DayOfWeek'])
X_test['Address'] = le.fit_transform(X_test['Address'])
X_test['Dates'] = le.fit_transform(X_test['Dates'])
```

```python
X_test = sc.transform(X_test)
```

```python
preds = model.predict(X_test)
preds
```

```python
output = pd.DataFrame({"Id":data_test.index, "Predictions":preds})
output.to_csv('submission.csv', index=False)
```
